```python
import pandas as pd

# Read the Excel file
file_path = 'FSImodified.xlsx'
df = pd.read_excel(file_path)

# Remove non-numeric characters from the "Rank" column
df['Rank'] = df['Rank'].str.replace(r'\D', '', regex=True)

# Convert the "Rank" column to a numeric data type
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce')

# Save the modified DataFrame back to the Excel file
df.to_excel(file_path, index=False)

# Now, the "Rank" column in FSImodified.xlsx contains only numeric values
```
