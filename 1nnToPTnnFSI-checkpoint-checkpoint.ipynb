```python
import os
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow import keras

# Define a function to remove ordinal suffixes (st, nd, rd) and convert to integer
def remove_ordinal_suffix(rank_str):
    try:
        return int(re.sub(r'\D', '', rank_str))
    except ValueError:
        return None  # Handle non-numeric or empty values


# Define a function to load and preprocess a single Excel file
def load_and_preprocess_excel(file_path):
    data = pd.read_excel(file_path)
    # Filter the columns
    data = data[['Country', 'Year', 'Rank']]
    # Remove ordinal suffix from Rank and convert to numeric
    data['Rank'] = data['Rank'].str.extract('(\d+)').astype(float)
    return data

# Define the folder path where the Excel files are located
folder_path = r"C:\Users\SANEABHUTTO\Desktop\FSI"

# Load and preprocess all Excel files in the folder
data_frames = []
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        data = load_and_preprocess_excel(file_path)
        data_frames.append(data)

# Concatenate data frames into one
combined_data = pd.concat(data_frames, ignore_index=True)

# Define a function to label data as politically stable or unstable
def label_political_stability(rank):
    return 1 if rank <= 75 else 0

# Apply the label_political_stability function to create the target variable
combined_data['Political_Stability'] = combined_data['Rank'].apply(label_political_stability)

# Split the data into training (2006-2018) and testing (2019-2023)
train_combined_data = combined_data[combined_data['Year'] <= 2018]
test_combined_data = combined_data[combined_data['Year'] >= 2019]

# Define features and target variable for training and testing
X_train = train_combined_data[['Year', 'Rank']].to_numpy()
X_test = test_combined_data[['Year', 'Rank']].to_numpy()
y_train = train_combined_data['Political_Stability'].to_numpy()
y_test = test_combined_data['Political_Stability'].to_numpy()

# ... (Previous code for loading and preprocessing data)

# Shuffle the training data to mix stable and unstable samples
X_train, y_train = shuffle(X_train, y_train, random_state=42)

# Calculate class weights
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)

# Define a neural network architecture with dropout for regularization
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),  # Input layer for Year and Rank
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),  # Dropout layer for regularization
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.2),  # Dropout layer for regularization
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model with a custom weighted loss function
def weighted_binary_crossentropy(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)  # Cast y_true to float32
    weights = (class_weights[1] / class_weights[0]) * (1 - y_true) + y_true
    return tf.keras.losses.binary_crossentropy(y_true, y_pred, from_logits=False) * weights

model.compile(optimizer='adam', loss=weighted_binary_crossentropy, metrics=['accuracy'])

# Train the model with class weights
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Fit the LabelEncoder on the training labels
label_encoder = LabelEncoder()
label_encoder.fit(y_train)

# Use the trained model to make predictions on the testing data
y_test_pred_probs = model.predict(X_test)
y_test_pred_binary = (y_test_pred_probs > 0.5).astype(int)

# Decode the binary predictions back to labels
y_test_pred_labels = label_encoder.inverse_transform(y_test_pred_binary.flatten())

# Print classification report and confusion matrix
print("Testing Classification Report:")
print(classification_report(y_test, y_test_pred_labels))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_labels))

# Access the original dataset and print the names of countries along with their predicted values
predicted_countries = test_combined_data[['Country']].copy()
predicted_countries['Predicted_Political_Stability'] = y_test_pred_labels
print("Countries with Predicted Political Stability:")
print(predicted_countries)

# ... (Continue with the rest of your code, e.g., making predictions for the year 2024)
```
