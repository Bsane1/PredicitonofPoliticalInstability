```python
import os
import pandas as pd
import re

# Define a function to remove ordinal suffixes (st, nd, rd) and convert to integer
def remove_ordinal_suffix(rank_str):
    try:
        return int(re.sub(r'\D', '', rank_str))
    except ValueError:
        return None  # Handle non-numeric or empty values

# Define the folder path where the Excel files are located
folder_path = r"C:\Users\SANEABHUTTO\Desktop\FSI"

# Initialize an empty list to store DataFrames
data_frames = []

# Loop through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(folder_path, filename)
        
        # Read the Excel file and select the required columns
        df = pd.read_excel(file_path, usecols=["Country", "Year", "Rank"])
        
        # Remove ordinal suffixes and convert "Rank" to integers, handle non-numeric or empty values
        df["Rank"] = df["Rank"].apply(remove_ordinal_suffix)
        
        # Filter countries based on their ranks and classify as politically unstable/stable
        df["Political_Stability"] = df["Rank"].apply(lambda x: "Politically Unstable" if (x is not None) and (x <= 75) else "Politically Stable")
        
        # Append the DataFrame to the list
        data_frames.append(df)

# Concatenate all DataFrames in the list into one DataFrame
combined_data = pd.concat(data_frames, ignore_index=True)

# Now, combined_data contains data from all the Excel files with the "Country," "Year," "Rank," and "Political_Stability" columns.
# You can manipulate and analyze the data as needed.

# Print all rows of the combined DataFrame (Note: Be cautious if the DataFrame is large)
#pd.set_option("display.max_rows", None)
#print(combined_data)

# Reset the option to its default value if needed
pd.reset_option("display.max_rows")
# Filter politically unstable countries
#politically_unstable_countries = combined_data[combined_data["Political_Stability"] == "Politically Unstable"]

# Iterate through the DataFrame and print politically unstable countries
#for index, row in combined_data.iterrows():
 #   if row["Political_Stability"] == "Politically Unstable":
  #      print(row)
# Filter politically unstable countries
politically_unstable_countries = combined_data[combined_data["Political_Stability"] == "Politically Unstable"]

# Iterate through the DataFrame and print politically unstable countries
for index, row in politically_unstable_countries.iterrows():
    print(row)

```
